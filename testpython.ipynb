```python
# function with argument and keywords

num1=int(input("enter number 1:"))
num2=int(input("enter number 2:"))
math_operate=input("type only these names: add, sub: ")

def bascic_cal(num1,num2,math_operate):
  if math_operate.lower()=="add":
    return num1+num2
  elif math_operate.lower()=="sub":
    return num1-num2
  else:
    print ("print correctly")

addition=bascic_cal(num1,num2,math_operate)
print(f"{addition}")




```
